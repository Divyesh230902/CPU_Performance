### Data Imported

### About Dataset

A dataset with 2,185 CPUs and 2,668 GPUs to help researchers understand the development trend of CPUs and GPUs.

Here are some interesting facts supported by the data:
- Moore's Law still holds, especially in GPUs.
- Dannard Scaling is still valid in general.
- CPUs have higher frequencies, but GPUs are catching up.
- GPU performance doubles every 1.5 years.
- GPU performance improvement is a joint effect of smaller transistors, larger die size, and higher frequency.
- High-end GPUs tends to first use new semiconductor technologies. Low-end GPUs may use old technologies for a few years.

```python
data = "chip_dataset.csv"
data_df = pd.read_csv(data)
```

```python
data_df = data_df.drop(columns=["Unnamed: 0"])
```

```python
data_df.value_counts()
```

```python
import numpy as np

# average process size

average_process_size = np.mean(data_df["Process Size (nm)"])
print("The Average process size is:", average_process_size)

# average TDP (Thermal Design Power)

average_tdp = np.mean(data_df["TDP (W)"])
print("The average Thermal Design Power is:", average_tdp)
```

```python
# Let's Take a summary of all the numerical data
data_summary = data_df[["Process Size (nm)", "TDP (W)", "Die Size (mm^2)", "Transistors (million)", "Freq (MHz)", "FP16 GFLOPS", "FP32 GFLOPS", "FP64 GFLOPS"]]
data_summary
```

```python
data_summary.describe()
```

```python
data_df_vendors = data_df["Vendor"]
data_df_vendors.unique()
```

# AMD Hardware

![AMD hardware](asset/th.jpeg)

```python
data_df_amd = data_df[(data_df["Vendor"] == 'AMD')]
data_df_amd
```

```python
data_df_amd.Type.value_counts()
```

```python
data_df_amd_summary = data_df_amd[["Process Size (nm)", "TDP (W)", "Die Size (mm^2)", "Transistors (million)", "Freq (MHz)", "FP16 GFLOPS", "FP32 GFLOPS", "FP64 GFLOPS"]].describe()
```

```python
data_df_amd_summary
```

```python
# get the high end AMD GPUs
data_df_amd_highend = data_df_amd[(data_df_amd["Type"] == 'GPU') & (data_df_amd["FP32 GFLOPS"] > 10000)]
data_df_amd_highend
```

```python
# get the high end AMD CPUs
data_df_amd_highend_cpu = data_df_amd[ (data_df_amd["Type"] == 'CPU') & (data_df_amd["Transistors (million)"] > 10000)]
data_df_amd_highend_cpu
```
